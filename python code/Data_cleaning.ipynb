```python
def missing_cols(df):
    '''prints out columns with its amount of missing values'''
    total = 0
    for col in df.columns:
        missing_vals = df[col].isnull().sum()
        total += missing_vals
        if missing_vals != 0:
            print(f"{col} => {df[col].isnull().sum()}")
    
    if total == 0:
        print("no missing values left")
            
missing_cols(data)
```

```python
# remove rows whose 标题 is missing
data.dropna(axis='index', how='any', subset=["标题"], inplace=True)
data = data.reset_index().drop("index",axis=1)
```

```python
# remove missing values 
data.dropna(axis='index', how='any', subset=["标题","设施", "标签", "位置"], inplace=True)
data = data.reset_index().drop("index",axis=1)
missing_cols(data)
```

```python
# find district
pattern_district = re.compile(r'市,(.*区)')

district = []
missing_id_list = []
num_missing = 0

for i in range(len(data)):
  match_district = pattern_district.findall(data["位置"][i])
  if match_district != []:
    l = match_district[0]
    pattern_onlyqu = re.compile(r'(.*?区)')
    onlyqu = pattern_onlyqu.findall(l)
    district.append(onlyqu[0].strip())
  else:
    num_missing +=1
    missing_id_list.append(i)
print(district)
print(num_missing)

# remove rows with no district information
data = data.drop(missing_id_list, axis='index')
data = data.reset_index().drop("index",axis=1)
data
```

```python
#find the list of labels
pattern_label = re.compile(r'\n(.*)')

labels = []

for i in range(len(data)):
  # labels
  match_label = pattern_label.findall(data["标签"][i])
  if match_label != []:
    labels.append(match_label)
  else:
    labels.append([])
print(labels)
```

```python
#find type
pattern_type = re.compile(r'·(.*)')

types = []

for i in range(len(data)):
  # labels
  match_type = pattern_type.findall(data["描述"][i])
  if match_type != []:
    types.append(match_type[0].strip())
  else:
    types.append([])
    print("error")
print(types)
```

```python
# find price
pattern_price = re.compile(r'￥(\d*.*\d*)/')

price = []

for i in range(len(data)):
  match_price = pattern_price.findall(data["价格"][i])
  if match_price != []:
    l = match_price[0].strip().translate(str.maketrans('', '', string.punctuation))
    price.append(l)
  else:
    print("error")
print(price)
```

```python
# find the number of facilities, grade and comments
pattern_bedroom = re.compile(r'(\d+(?:\.\d+)?) 间卧室')
pattern_bed = re.compile(r'(\d+(?:\.\d+)?).*床')
pattern_toilet = re.compile(r'(\d+(?:\.\d+)?).*卫生间')  
pattern_people = re.compile(r'可住.*(\d+(?:\.\d+)?)') 
pattern_comment = re.compile(r'(\d+(?:\.\d+)?)条评论')
pattern_grade = re.compile(r'(\d+(?:\.\d+)?)分')

num_bedrooms = []
num_toilet = []
num_bed = []
num_people = []
num_comment = []
grade = []

num_new1 = 0
num_new2 = 0

for i in range(len(data)):
  # bedrooms
  match_bedroom = pattern_bedroom.findall(data["设施"][i])
  if match_bedroom != []:
    num_bedrooms.append(match_bedroom[0])
  else:
    num_bedrooms.append("0")

  # beds
  match_bed = pattern_bed.findall(data["设施"][i])
  if match_bed != []:
    num_bed.append(match_bed[0])
  else:
    num_bed.append("0")

  # 卫生间
  match_toilet = pattern_toilet.findall(data["设施"][i])
  if match_toilet != []:
    num_toilet.append(match_toilet[0])
  else:
    num_toilet.append("0")

  # 可住人数
  match_people = pattern_people.findall(data["设施"][i])
  if match_people != []:
    num_people.append(match_people[0])
  else:
    num_people.append("0")
  
  #评论数
  match_comment = pattern_comment.findall(data["评分/评价"][i])
  if match_comment != []:
    num_comment.append(match_comment[0])
  else:
    num_comment.append(np.nan)
    num_new1 += 1

  #评分
  match_grade = pattern_grade.findall(data["评分/评价"][i])
  if match_grade != []:
    grade.append(match_grade[0])
  else:
    grade.append(np.nan)
    num_new2 += 1
  

print("num_bedrooms: ",num_bedrooms)
print("num_bed: ",num_bed)
print("num_toilet: ", num_toilet)
print("num_people: ", num_people)
print("num_comment: ", num_comment)
print("grade: ", grade)

print(num_new1)
print(num_new2)
```

```python
df = pd.DataFrame([data["标题"].tolist(), num_comment, grade, price, district, types, num_bedrooms, num_bed, num_toilet, num_people, labels]).T
df.columns = ["title","num_comments","grade","price","district","type","num_bedrooms","num_beds","num_toilets","num_people_can_in","labels"]
df = df.dropna()
```
