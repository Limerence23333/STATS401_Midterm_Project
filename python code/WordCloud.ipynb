## Dealing with Titles

```python
spe_chars = ['“','”','｛','｝','[',']','［','］','✔','；','〖','〗','☆','#','★','〔','〕','┃','','@','+',':','•','*','.','／','/','【','】','-','|','｜','『','』',' ','I','「','」','、',',','～','（','）','❤','♥','，','！','.','~','－','》','《','(',')','·','。','…','&','}','{','：','丨']
def remove_spe(s):
    cleaned = s
    for c in spe_chars:
        cleaned = cleaned.replace(c,'')
    return cleaned
high_pop = list(top200['title'])
for i in range(len(high_pop)):
    high_pop[i] = remove_spe(high_pop[i])
```

```python
high_pop
```

```python
import jieba as jb
for i in range(len(high_pop)):
    seg = jb.cut(high_pop[i])
    high_pop[i] = " ".join(seg)
high_pop_df = pd.DataFrame(high_pop)
```

```python
high_pop_df.to_csv('sh_top200.csv')
```

# XIAMEN

```python
df = pd.read_excel('爱彼迎厦门.xlsx')
df = df.dropna()

keep_index = []
for i in range(len(df.index)):
    if df.iloc[i,3][0] in '012345' and df.iloc[i,3][1] == '.': 
        keep_index.append(i)
len(keep_index)

df_target = df.iloc[keep_index,:]
```

```python
def get_score(val):
    score = int(val[0]) + int(val[2])*0.1
    return score
def get_comment(val):
    return int(val[7])

scores = list(df_target['评分/评价'].map(get_score))
num_comments = list(df_target['评分/评价'].map(get_comment))

df_target['score'] = scores
df_target['num_comment'] = num_comments
df_target["popularity"] = df_target["num_comment"]*np.log(df_target["score"])
```

```python
df_target = df_target.reset_index()
df_target=df_target.drop(['index','level_0'],axis=1)
df_target
```

```python
# top200
top200 = df_target.sort_values(by='popularity',ascending=False)[:200]
top200
```

```python
spe_chars = ['［','］','✔','；','〖','〗','☆','#','★','〔','〕','┃','','@','+',':','•','*','.','／','/','【','】','-','|','｜','『','』',' ','I','「','」','、',',','～','（','）','❤','♥','，','！','.','~','－','》','《','(',')','·','。','…','&','}','{','：','丨']
def remove_spe(s):
    cleaned = s
    for c in spe_chars:
        cleaned = cleaned.replace(c,'')
    return cleaned
high_pop = list(top200['标题'])
for i in range(len(high_pop)):
    high_pop[i] = remove_spe(high_pop[i])
```

```python
import jieba as jb
jb.add_word("曾厝垵", freq = 20000, tag = None)
jb.add_word("沙坡尾", freq = 20000, tag = None)
for i in range(len(high_pop)):
    seg = jb.cut(high_pop[i])
    high_pop[i] = " ".join(seg)
high_pop_df = pd.DataFrame(high_pop,columns=['title'])
high_pop_df.to_csv('xm_top200.csv')
```
